```python
base_url = "https://www2.hm.com"

sale_prices = []
original_prices = []
categories_list = []
sub_category_list = []
product_ids = []
product_names = []
product_links = []
product_images = []

j = 0
for category in category_info: 
    j = j+1
    if j == 1: continue
    print(category['Category'])
    json_response = get_json_response(category['Link'])
    if json_response != None: 
        for product in json_response['products']:
            print([product['title']])
            categories_list.append(category['Gender'] + ' ' + category['Category'])
            sub_category_list.append(category['Sub Category'])
            product_ids.append(product['articleCode'])
            product_links.append(base_url + product['link'])
            if 'image' in product: product_images.append([product['image'][0]['src'].lstrip('//'),
                                                        product['image'][0]['dataAltImage'].lstrip('//')])
            else : product_images.append([])
            sale_prices.append('')
            original_prices.append(product['price'].lstrip('Rs.').rstrip('.00'))
            product_names.append(product['title'])
```

```python
import pandas as pd

df = pd.DataFrame({
    'Category': categories_list,
    'Sub Category': sub_category_list,
    'Product Name': product_names,
    'Product Link': product_links,
    'Sale Price': sale_prices,
    'Original Price': original_prices,
    'Primary Image': product_images,
    'Product Secondary Images': product_images,
    'Product Id': product_ids
})

df.to_csv('h&m.csv', index=False)
print("Data scraped and saved successfully!")
```
