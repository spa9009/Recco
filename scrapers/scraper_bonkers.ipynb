```python
import pandas as pd
import requests as r
from bs4 import BeautifulSoup
```

```python
url_list = ["https://www.bonkerscorner.com/collections/oversized-tshirt-men",
       "https://www.bonkerscorner.com/collections/bottoms-mens",
       "https://www.bonkerscorner.com/collections/sweatshirts-hoodies-men",
       "https://www.bonkerscorner.com/collections/basic-men",
       "https://www.bonkerscorner.com/collections/bottoms-women",
       "https://www.bonkerscorner.com/collections/oversized-tshirt-women",
       "https://www.bonkerscorner.com/collections/tops-women",
       "https://www.bonkerscorner.com/collections/sweatshirt-hoodie-women",
       "https://www.bonkerscorner.com/collections/basics-women",
       "https://www.bonkerscorner.com/collections/jacket-women",
       "https://www.bonkerscorner.com/collections/dress",
       "https://www.bonkerscorner.com/collections/signature-women",
       "https://www.bonkerscorner.com/collections/oversized-shirt",
       "https://www.bonkerscorner.com/collections/regular-fit-shirt-men",
       "https://www.bonkerscorner.com/collections/signature-men",
       "https://www.bonkerscorner.com/collections/jeans-men",
       "https://www.bonkerscorner.com/collections/polo-men",
       "https://www.bonkerscorner.com/collections/caps",
       "https://www.bonkerscorner.com/collections/supima-men",
       "https://www.bonkerscorner.com/collections/seamless",
       "https://www.bonkerscorner.com/collections/corduroy",
       "https://www.bonkerscorner.com/collections/polyamide-collection"
       ]
```

```python
product_names = []
sale_prices = []
original_prices = []
product_primary_images = []
product_seconday_images = []
sizes = []
categories = []
product_descriptions = []
product_urls = []

base_url = "https://www.bonkerscorner.com"

denied_child = ["www.bonkerscorner.com/cdn/shop/files/Bonkers_Corner_Logo_White1_20x4_crop_center.png?v=1721047270",
"www.bonkerscorner.com/cdn/shop/files/KK_1_20x_crop_center.jpg?v=1720189234",
"www.bonkerscorner.com/cdn/shop/files/20230601_112131826_iOS_20x_crop_center.jpg?v=1720186984",
"www.bonkerscorner.com/cdn/shop/files/Bonkerscorner_loi_03_20x_crop_center.jpg?v=1720185860",
"www.bonkerscorner.com/cdn/shop/files/Bonkerscorner_Chainsaw-men_05_20x_crop_center.jpg?v=1720184478",
"www.bonkerscorner.com/cdn/shop/files/Bonkerscorner_Slayer_04_20x_crop_center.jpg?v=1720184335",
"www.bonkerscorner.com/cdn/shop/files/Bonkerscorner_akira_06_20x_crop_center.jpg?v=1720184066",
"www.bonkerscorner.com/cdn/shop/files/IMG_9671_20x_crop_center.jpg?v=1723785516"]

for url in url_list:
    req = r.get(url)
    soup = BeautifulSoup(req.text, 'html.parser')

    product_containers = soup.find_all('product-card', class_='product-card text-left')
    # print(product_containers[0])
    for product in product_containers : 
        product_names.append(product.find('a', class_='product-card-title').text.strip())
        print(product.find('a', class_='product-card-title').text.strip())
        product_primary_images.append(product.find('img', class_='lazyload product-primary-image').get('src').lstrip('//'))
        categories.append(soup.find('h1', class_='collection-banner--title').text.strip())
        price_info = product.find('span', class_='price').text.split('.')
        original_prices.append(price_info[1].strip())
        sale_prices.append(price_info[3].strip())

        child_url = base_url + product.find('a', class_='product-featured-image-link aspect-ratio aspect-ratio--adapt').get('href').strip()
        product_urls.append(child_url)
        child_req = r.get(child_url)
        child_soup = BeautifulSoup(child_req.text, 'html.parser')
        # print(child_soup)
        child_img_soups = child_soup.find_all('img', class_='lazyload')
        sec_images = []
        for soupc in child_img_soups:
            if soupc.get('src') != None: 
                urlx = soupc.get('src').lstrip('//')
                if urlx not in denied_child: sec_images.append(urlx) 
        product_seconday_images.append(sec_images)
        if child_soup.find('div', class_='p-rich_text_section'):
            product_descriptions.append(child_soup.find('div', class_='p-rich_text_section').text.strip().replace('\n', ' ').strip())
        else :
            product_descriptions.append('')
        
        size_array = []
        if child_soup.find('div', class_='variations'):
            for soups in child_soup.find('div', class_='variations').find_all('span'):
                size_array.append(soups.text.strip())
        sizes.append(size_array)


# df = pd.DataFrame({
#     'Product Name': product_names,
#     'Price': prices,
#     'Variant': item_variants,
#     'Category': item_list_names,
#     'ImageURL': image_urls
# })


df = pd.DataFrame({
    'Category': categories,
    'Product Name': product_names,
    'Product Link': product_urls,
    'Sale Price': sale_prices,
    'Original Price': original_prices,
    'Sizes': sizes,
    'Primary Image': product_primary_images,
    'Product Secondary Images': product_seconday_images,
    'Product URLs': product_urls,
    'Product Description': product_descriptions
})

df.to_csv('bonkers.csv', index=False)
print("Data scraped and saved successfully!")
```
