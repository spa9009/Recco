```python
import pandas as pd
import requests as r
from bs4 import BeautifulSoup
```

```python
url = "https://www.urbanmonkey.com/collections/hoodies"

req1 = r.get(url)

soup = BeautifulSoup(req1.text, 'html.parser')
```

```python
print(soup)
```

```python
# Define the set of common classes you're looking for
common_classes = {'gtm-product-card', 'gtm-product-info', 't4s-pr-grid', 't4s-pr-style6'}

    # Find all divs where the class attribute contains all of the common classes
product_containers = soup.find_all('div', class_=lambda x: x and common_classes.issubset(x.split()))
```

```python
print(product_containers[0])
```

```python
product_containers[1].find('h3').text
```

```python
import re
import json

def is_valid_json(json_string):
    try:
        json.loads(json_string)
        return True
    except json.JSONDecodeError:
        return False
    
product_names = []
prices = []
item_variants = []
image_urls = []
item_list_names = []

for product in product_containers :
    product_name = product.find('h3').text.strip()
    # print(product_name)
    image_link = product.find('img', class_='t4s-product-main-img')
    image_url = image_link.get('data-src') or image_link.get('src')
    print(image_url)
    product_div = product.get('data-layer-product-info', 'N/A')
    if is_valid_json(product_div) :
        json_product = json.loads(product_div)
        item_variant = json_product.get('item_variant', 'N/A')
        item_list_name = json_product.get('item_list_name', 'N/A')
        price = json_product.get('price', 'N/A')
    else :
        item_variant = ''
        item_list_name = ''
        price = ''
    product_names.append(product_name)
    prices.append(price)
    item_variants.append(item_variant)
    image_urls.append(image_url)
    item_list_names.append(item_list_name)


df = pd.DataFrame({
    'Product Name': product_names,
    'Price': prices,
    'Variant': item_variants,
    'Category': item_list_names,
    'ImageURL': image_urls
})

df.to_csv('product_data7.csv', index=False)
print("Data scraped and saved successfully!")
```

```python
import glob
# Merge CSVs

csv_files = csv_files = glob.glob('*.csv')

# Read the first CSV file into a DataFrame
merged_df = pd.DataFrame()

for file in csv_files:
    df = pd.read_csv(file)
    # Reindex DataFrame to ensure all columns are present
    df = df.reindex(columns=merged_df.columns)
    # Append the DataFrame to the merged DataFrame
    merged_df = pd.concat([merged_df, df], ignore_index=True)

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('merged_output.csv', index=False)

print("CSV files merged successfully with missing columns handled!")
```
